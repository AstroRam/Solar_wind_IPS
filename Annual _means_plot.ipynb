```python
import numpy as np
import pandas as pd
data = pd.read_csv('annualmeans-scindex.txt', sep ='\s+', header=None)
x=data[0]
y=data[1]
```

```python
import matplotlib.pyplot as plt
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
plt.xlabel('Years')
plt.ylabel('scintillation index')
plt.title('Temporal Variation of Scintillation Index')
plt.scatter(x,y)
```

```python
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
results = sm.OLS(y,sm.add_constant(x)).fit()
plt.scatter(x,y)
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)),'g')
plt.show()
```
